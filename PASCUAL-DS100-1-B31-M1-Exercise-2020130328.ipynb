```python
import requests as rq
from io import BytesIO

import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/Ackerman407/DS100/main/supermarket_sales%20-%20Sheet1.xlsx'
dt = rq.get(url).content
data = pd.read_excel(BytesIO(dt))
pd.set_option('display.max_rows',None)
data.head()
```

```python
#selecting the Cities Yangon and Mandalay only
city_data = data.loc[data['City'].str.contains('Yangon|Mandalay')]
city_data
```

```python
city_data = city_data.loc[city_data['Product line'].str.contains('Fashion accessories|Health and beauty')]
city_data
```

```python
city_data = city_data[city_data['Total'] >= 600]
city_data
```

```python
city_data['Avg. Tax per Product'] = city_data['Tax 5%'] / city_data['Quantity']
city_data.update(city_data['Avg. Tax per Product'].fillna(0))
city_data
```

```python
city_data['Actual Cost per Product'] = city_data['cogs'] / city_data['Quantity']
city_data
```

```python
df1=city_data.loc[city_data['City'].str.contains('Yangon')]
df2=city_data.loc[city_data['City'].str.contains('Mandalay')]

df1=(df1.rename(columns={'City': 'City A','Avg. Tax per Product':'Avg. Tax per Product 1','Actual Cost per Product':'Actual Cost per Product 1'}))
df2=(df2.rename(columns={'City': 'City B','Avg. Tax per Product':'Avg. Tax per Product 2','Actual Cost per Product':'Actual Cost per Product 2'}))

df1=df1.drop(columns=['Invoice ID','Branch','Customer type','Gender','Product line','Unit price','Quantity','Tax 5%','Total','Time','Date','cogs','gross margin percentage','gross income','Rating'])
df2=df2.drop(columns=['Invoice ID','Branch','Customer type','Gender','Product line','Unit price','Quantity','Tax 5%','Total','Time','Date','cogs','gross margin percentage','gross income','Rating'])
abs_data=pd.merge(df1,df2)

abs_data[['City A','Avg. Tax per Product 1','Actual Cost per Product 1','City B','Avg. Tax per Product 2','Actual Cost per Product 2']]
```

```python
abs_data['Cost Difference between Yangon and Mandalay City']=np.round(abs(abs_data['Actual Cost per Product 1']-abs_data['Actual Cost per Product 2']),3)
abs_data[['City A','Avg. Tax per Product 1','Actual Cost per Product 1','City B','Avg. Tax per Product 2','Actual Cost per Product 2','Cost Difference between Yangon and Mandalay City']]
```
